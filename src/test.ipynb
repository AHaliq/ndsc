## Add category group column to data

```python
import pandas as pd

cat = pd.read_json('./data/source/categories.json')
cgmap = {}
for i in cat['Mobile'].unique():
    cgmap[i] = 0
for i in cat['Fashion'].unique():
    cgmap[i] = 1
for i in cat['Beauty'].unique():
    cgmap[i] = 2
# get group from category

data = pd.read_csv("./data/source/train.csv")
data['group'] = data.apply(lambda row: cgmap[row['Category']], axis=1)
# add new column

data.to_csv('./data/clean/train_group.csv')
# export
```

```python
from PIL import Image
import requests
from io import BytesIO
import pandas as pd

IMAGES = 'http://storage.googleapis.com/ndsc-storage/data/images/'

data = pd.read_csv("./data/clean/train_group.csv")
imgs = []
for i in range(5):
    response = requests.get(IMAGES + data['image_path'][i])
    imgs[i] = Image.open(BytesIO(response.content))
from matplotlib.pyplot import imshow
import numpy as np
imshow(np.asarray(imgs))
```

```python
bucket = 'http://storage.googleapis.com/ndsc-storage/'
data = bucket + 'data/'
images = data + 'images/'
source = data + 'source/'
clean = data + 'clean/'

import pandas as pd
import re

def data_addgroup(d):
    def _strtogroup(s):
        if re.search('^mobile', s):
            return 0
        elif re.search('^fashion', s):
            return 1
        return 2
    d['group'] = d.apply(lambda row: _strtogroup(row['image_path']), axis=1)
    return d

import pandas as pd
d = pd.read_csv(source + 'test.csv')
data_addgroup(d)
d.head()
```
